```python
#a. Build 99% Confidence Interval Using Sample Standard Deviation
```

```python
import numpy as np
from numpy import*
import pandas as pd
from pandas import*
import scipy.stats as stats
import math
```

```python
data=(1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29)
sample=DataFrame(data)
sample
```

```python
#sample size
len(sample)
```

```python
sample_mean=sample.mean()
print('The mean of the sample is :',sample_mean)
print('\n')
sample_std=sample.std()
print('The standard deviation of the sample is :',sample_std)
print('\n')
t_critical=stats.t.cdf(x=0.99,df=14)
print('The t-critical value is :',t_critical)
print('\n')
margin_of_error=t_critical*(sample.std()/sqrt(len(sample)))
confidence_interval=(sample_mean-margin_of_error,sample_mean+margin_of_error)
print('The 99% of confidence interval is :',confidence_interval)
```

The sample size is smaller and population standard deviation is not present,Thats why here t distribution is used

```python
sample_mean=sample.mean()
print('The Mean Of The Sample Is :',sample_mean)
print('\n')
z_critical=stats.norm.cdf(x=0.99)
print('The Z-Critical Value Is :',z_critical)
print('\n')
pop_stdev=0.2
margin_of_error=z_critical*(pop_stdev/math.sqrt(len(sample)))
confidence_interval=(sample_mean-margin_of_error,sample_mean+margin_of_error)
print('The 99% Of Confidence Interval Is :',confidence_interval)
```
