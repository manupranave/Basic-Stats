#  Standardization of Numerical Variables

```python
#Explaination of the concept of standardization (z-score normalization):
```

Standardization, also known as z-score normalization, is a technique used in statistics and data analysis to transform data such that it has a mean of 0 and a standard deviation of 1. This process allows data from different scales to be compared directly by making the distribution of the data more consistent.

```python
data
```

```python
numerical_columns=data.select_dtypes(include=['int64','float64'])
```

```python
mean=data.describe().mean()
std=data.describe().std()

for column in numerical_columns:
  data_standardized=(data[column]-mean[column])/std[column]
print('Standardized DataFrame:')
print(data)
```

```python
for col in numerical_columns:
  figure(figsize=(10,6))
  subplot(1,2,1)
  data[col].plot(kind='hist',bins=50)
  title('Before Standardization')
  xlabel(col)
  ylabel('Frequency')

  subplot(1,2,2)
  data_standardized.plot(kind='hist',bins=50)
  title('After Standardization')
  xlabel(col + 'Standardized')
  ylabel('Frequency')
  show()
```

# **Conversion of Categorical Data into Dummy Variables**

One-hot encoding is a technique to convert categorical data into binary vectors, where each category is represented by a binary column (0 or 1). This is necessary because many machine learning algorithms cannot handle categorical data directly.

```python
#	Apply one-hot encoding to the categorical columns
sales_data_onehot=get_dummies(data,columns=categorical_columns)
sales_data_onehot
```

```python
#creating binary (0 or 1) columns for each category.
sales_data_transformed=concat([data_standardized,sales_data_onehot.drop(columns=numerical_columns)],axis=1)
sales_data_transformed
```
